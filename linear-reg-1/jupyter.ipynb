```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#CODE -1 Data Generation
def generateData(m):
  X=np.random.randn(m)*10;
  noise = np.random.randn(m)
  y=3*X + 1 + 5*noise
  return X,y

X,y=generateData(100)
plt.scatter(X,y)
plt.show()
```

```python
#Plotting the Data
def plotData(X,y , color="orange",title="Data"):
  plt.title(title)
  plt.xlabel("X")
  plt.ylabel('y')
  plt.scatter(X,y,c=color)
  plt.show()
plotData(X,y)
```

```python
#CODE -2 Data Normalisation
def Normalisation(X):
  X=(X-X.mean())/X.std()
  return X
X=Normalisation(X)
plotData(X,y)
```

```python
#CODE -3 Train And Test Data
def TrainTestSplit(X,y,split=0.8):
  m=X.shape[0];
  data= np.zeros((m,2))
  data[:,0]=X;
  data[:,1]=y;
  split=int(m*split)
  #shuffling the data
  np.random.shuffle(data)
  XT = data[:split , 0]
  yT = data[:split , 1]
  Xt = data[split: , 0]
  yt = data[split: , 1]
  return XT,yT,Xt,yt
XT,yT,Xt,yt=TrainTestSplit(X,y)
plotData(XT,yT )
plotData(Xt,yt )
```
